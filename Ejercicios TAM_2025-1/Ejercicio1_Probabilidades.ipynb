```python
import numpy as np

# Función para obtener las probabilidades del usuario
def obtener_probabilidades():
    while True:
        try:
            n = int(input("Ingrese el número de cajas: "))  # Número de cajas
            m = int(input("Ingrese el número de frutas: "))  # Número de frutas
            if n <= 0 or m <= 0:
                print("Por favor, ingrese un número positivo para el número de cajas y frutas.")
                continue
            break
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número entero.")

    # Ingresar las probabilidades de las cajas (P(B = r), P(B = b), ...)
    P_B = np.zeros(n)
    for i in range(n):
        while True:
            try:
                P_B[i] = float(input(f"Ingrese la probabilidad de seleccionar la caja {i+1} (P(B = {i+1})): "))
                if P_B[i] < 0 or P_B[i] > 1:
                    print("La probabilidad debe estar entre 0 y 1. Inténtelo de nuevo.")
                    continue
                break
            except ValueError:
                print("Entrada no válida. Ingrese un valor numérico entre 0 y 1.")

    # Validar que las probabilidades a priori sumen 1
    if abs(sum(P_B) - 1) > 1e-6:
        print("Error: Las probabilidades a priori de las cajas no suman 1.")
        return None, None, None, None

    # Ingresar la cantidad de frutas por tipo en cada caja
    frutas_por_caja = []
    for i in range(n):
        frutas = []
        print(f"Ingrese la cantidad de frutas en la caja {i+1}:")
        for j in range(m):
            while True:
                try:
                    cantidad = int(input(f"Cantidad de fruta {j+1} en la caja {i+1}: "))
                    if cantidad < 0:
                        print("La cantidad de frutas no puede ser negativa. Inténtelo de nuevo.")
                        continue
                    frutas.append(cantidad)
                    break
                except ValueError:
                    print("Entrada no válida. Ingrese un número entero.")
        frutas_por_caja.append(frutas)

    return P_B, frutas_por_caja, n, m

# Función para calcular las probabilidades condicionales automáticamente
def calcular_probabilidades_condicionales(frutas_por_caja, n, m):
    P_F_given_B = np.zeros((n, m))

    for i in range(n):  # Para cada caja
        total_frutas = sum(frutas_por_caja[i])  # Total de frutas en la caja
        for j in range(m):  # Para cada tipo de fruta
            P_F_given_B[i][j] = frutas_por_caja[i][j] / total_frutas  # Probabilidad condicional

    return P_F_given_B

# Función para calcular las probabilidades totales P(F)
def probabilidad_total(P_F_given_B, P_B, n, m):
    P_F = np.zeros(m)
    for j in range(m):  # Para cada fruta
        P_F[j] = sum(P_F_given_B[i][j] * P_B[i] for i in range(n))
    return P_F

# Función para calcular las probabilidades condicionales usando el Teorema de Bayes
def bayes(P_F_given_B, P_B, P_F, n, m):
    P_B_given_F = np.zeros((n, m))

    for j in range(m):  # Para cada fruta
        for i in range(n):  # Para cada caja
            P_B_given_F[i][j] = (P_F_given_B[i][j] * P_B[i]) / P_F[j]

    return P_B_given_F

# Función principal para obtener la solución
def main():
    # Paso 1: Obtener las probabilidades del usuario
    P_B, frutas_por_caja, n, m = obtener_probabilidades()

    # Verificar si hubo un error en la obtención de probabilidades
    if P_B is None or frutas_por_caja is None:
        return

    # Paso 2: Calcular las probabilidades condicionales automáticamente
    P_F_given_B = calcular_probabilidades_condicionales(frutas_por_caja, n, m)

    # Paso 3: Calcular las probabilidades totales P(F)
    P_F = probabilidad_total(P_F_given_B, P_B, n, m)

    # Paso 4: Aplicar el Teorema de Bayes para obtener P(B | F)
    P_B_given_F = bayes(P_F_given_B, P_B, P_F, n, m)

    # Mostrar los resultados
    print("\n### Probabilidades Totales (P(F)) ###")
    for j in range(m):
        print(f"Fruta {j+1}: P(F = {j+1}) = {P_F[j]}")

    print("\n### Probabilidades Condicionales (P(B | F)) ###")
    for j in range(m):
        print(f"\nPara la fruta {j+1}:")
        for i in range(n):
            print(f"P(B = {i+1} | F = {j+1}) = {P_B_given_F[i][j]}")

    print("\n### Prioris de las Cajas (P(B)) ###")
    for i in range(n):
        print(f"P(B = {i+1}) = {P_B[i]}")

    print("\n### Verosimilitudes (P(F | B)) ###")
    for i in range(n):
        print(f"\nPara la caja {i+1}:")
        for j in range(m):
            print(f"P(F = {j+1} | B = {i+1}) = {P_F_given_B[i][j]}")

# Ejecutar el programa
main()

```
