## Revisi칩n general del dataset

El primer paso de la exploraci칩n consiste en comprender la estructura b치sica del conjunto de datos: n칰mero de filas y columnas, nombres de las variables, tipos de datos, y estad칤sticas descriptivas generales.

Este diagn칩stico inicial permitir치 detectar posibles problemas como valores nulos, tipos de datos incorrectos, presencia de duplicados, o columnas irrelevantes para la tarea de aprendizaje supervisado.


```python
import pandas as pd

# cARGAR Dataset
df = pd.read_csv("data/AmesHousing.csv")


# Dimensiones del dataset
print("Dimensiones del dataset:", df.shape)

# Nombres de columnas
print("\nColumnas del dataset:")
print(df.columns.tolist())
```

```python
# Informaci칩n general del DataFrame
df.info()
```

```python
# Estad칤sticas b치sicas
df.describe().T
```

```python
# Ver cu치ntas variables categ칩ricas hay
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
print(f"\nVariables categ칩ricas ({len(categorical_cols)}):", categorical_cols)

# Frecuencias para las primeras 5 columnas categ칩ricas
for col in categorical_cols[:5]:
    print(f"\nFrecuencias de la variable '{col}':")
    print(df[col].value_counts(dropna=False).head())
```

```python
# Conteo de valores nulos por columna
null_counts = df.isnull().sum()
null_counts = null_counts[null_counts > 0].sort_values(ascending=False)
print("\nColumnas con valores nulos:")
print(null_counts)
```

```python
# Verificar si hay filas duplicadas
print("\nN칰mero de filas duplicadas:", df.duplicated().sum())
```
